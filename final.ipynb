```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from IPython.display import HTML
```

```python
df= pd.read_csv("VISUA_FINAL.csv",thousands=',')
```

```python
delitos= df.iloc[:,:]
delitos["crimenes"]= delitos.iloc[:, 1:4].sum(axis=1)
top_10=delitos.nlargest(15,'crimenes')
```

```python
top_10
```

```python
top_10["id_col"] = top_10.groupby(['Colonia']).ngroup()
top_10["poblacion"] =top_10["Población femenina"]+top_10["Población masculina"]
escuela=top_10.sort_values(by=["Población que no asiste a la escuela"],ascending=True)
trabajo=top_10.sort_values(by=["Población de 12 años y más no económicamente activa"],ascending=True)
```

```python
top_10.nlargest(10,'poblacion')
```

```python
escuela["Población que no asiste a la escuela"] = pd.to_numeric(escuela["Población que no asiste a la escuela"])
escuela.sort_values(by = ['Población que no asiste a la escuela'], ascending = True, inplace=True)
trabajo["Población de 12 años y más no económicamente activa"] = pd.to_numeric(trabajo["Población de 12 años y más no económicamente activa"])
trabajo.sort_values(by = ['Población de 12 años y más no económicamente activa'], ascending = True, inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.set_xlabel('Población que no asiste a la escuela', fontsize=15)
ax.set_ylabel('crimenes', fontsize=15)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
sns.scatterplot(data=escuela, y="crimenes", x="Población que no asiste a la escuela",hue="Colonia",size="crimenes",sizes=(100,200))
plt.savefig("educacion.png")
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.set_xlabel('Población que no trabaja', fontsize=15)
ax.set_ylabel('crimenes', fontsize=15)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
sns.scatterplot(data=trabajo, y="crimenes", x="Población de 12 años y más no económicamente activa",hue="Colonia",size="crimenes",sizes=(100,200))
plt.savefig("no_trabajo.png")
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.set_xlabel('Población que no trabaja', fontsize=15)
ax.set_ylabel('crimenes', fontsize=15)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
sns.barplot(data=top_10[:5], x="poblacion",y="crimenes",hue="Colonia")
plt.savefig("poblacion.png")
```
